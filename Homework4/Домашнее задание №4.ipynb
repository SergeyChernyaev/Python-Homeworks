````python
# Задание № 1

# Пункт 1. Пользователь по команде "p" может узнать владельца документа по его номеру
def show_onwer(iterable, value):
    str = 'Документ не найден в базе'
    for  dict_ in iterable:
        if dict_['number'] == value:
            str = 'Владелец документа: ' + dict_['name']
            break
    return (str)

#Пункт 2. Пользователь по команде "s" может по номеру документа узнать на какой полке он хранится
def show_shelf(iterable, value):
    key = None
    for key, list_ in iterable.items():
        for doc_num in list_:
            if doc_num == value:
                return (key) # вернуть номер полки

#Пункт 3. Пользователь по команде "l" может увидеть полную информацию по всем документам
def show_all_docs (iterable_1,iterable_2):
    print('Текущий список документов:')
    for  dict_ in iterable_1:
        shelf = show_shelf(iterable_2, dict_['number'])
        print(f"№: {dict_['number']}, тип: {dict_['type']}, владелец: {dict_['name']}, полка хранения:{shelf}")
    return ()
  
# Пункт 4. Пользователь по команде "ads" может добавить новую полку      
def add_shelf(iterable,num_shelf):
    if num_shelf not in iterable.keys():
        iterable[num_shelf] = []
        print ('Полка добавлена.', end = '')
    else:
        print('Такая полка уже существует. ', end = '')
# распечатать список полок        
    print_shelfs(iterable)
    return ()

# Функция печати списка полок:
def print_shelfs(iterable):
    shelfs_list = []
    for key in iterable.keys(): # можно Лямбдой
        shelfs_list.append(key) 
    print ('Текущий перечень полок: ', ',' .join(shelfs_list))
    return () 

# Пункт 5. Пользователь по команде "ds" может удалить существующую полку из данных (только если она пустая)
def del_shelf(iterable, num_shelf):
    if num_shelf  in iterable.keys():
        if iterable[num_shelf] == []:
            iterable.pop(num_shelf)
            print ('Полка удалена.', end = '')
        else:
            print ('На полке есть документы, удалите их перед удалением полки.', end = '')
    else:
        print('Такой полки не существует.', end = '')
# распечатать список полок        
    print_shelfs(iterable)
    return ()

# Задание № 2 (дополнительное)

# Пункт 1. Пользователь по команде "ad" может добавить новый документ в данные
def add_doc(iterable_1, iterable_2):
    doc_num = input('Введите номер документа: ')
    doc_type = input('Введите тип документа: ')
    doc_onwer = input('Введите владельца документа: ')
    num_shelf = input('Введите полку для хранения: ')
    if num_shelf not in iterable_2.keys():
        print ('Такой полки не существует.Добавьте полку командой as.')
    else:
        iterable_1.append({'type': doc_type, 'number': doc_num, 'name': doc_onwer}) # добавить словарь
        iterable_2[num_shelf].append(doc_num) # добавить номер документа в список на полку
        print('Документ добавлен.')
        show_all_docs (iterable_1, iterable_2) 
    return()


# Пункт 2. Пользователь по команде "d" может удалить документ из данных
def del_doc(iterable_1, iterable_2):
    doc_num = input('Введите номер документа: ')
    
    str = 'Документ не найден в базе'
    flag = True
    
    for  index, dict_ in enumerate(iterable_1):
        if dict_['number'] == doc_num:
            # Удалить словарь
            iterable_1.pop(index)
            # Удалить документ с полки
            for list_ in iterable_2.values():
                if flag: # не крутить лишние циклы
                    for val  in list_:  # смотрим список
                        if val == doc_num: # пока не нашли документ на полке
                            list_.remove(doc_num)
                            flag = False
                            str = 'Документ удален'
                            break
                else:
                    break
    print (str)        
    show_all_docs (iterable_1,iterable_2) 
    return()

#Пункт 3. Пользователь по команде "m" может переместить документ с полки на полку

def move_doc(iterable_1, iterable_2):
    doc_num = input('Введите номер документа: ')
    shelf_num = input('Введите номер полки: ')
    shelf_now = show_shelf(iterable_2, doc_num) # сейчас документ лежит на полке..
    
    if shelf_now == None: 
        print('Документ не найден в базе.')
        show_all_docs (iterable_1,iterable_2)
    if shelf_now == shelf_num:
        print(f'Документа {doc_num} находится на полке {shelf_now}. Перемещение не требуется.')
    else:
        if shelf_num not in iterable_2.keys():
            print('Такой полки не существует. Текущий перечень полок:', end='')
            print_shelfs(iterable_2)
        else:
            iterable_2[shelf_num].append(doc_num)   
            iterable_2[shelf_now].remove(doc_num)
            print(f'Документ {doc_num} перемещен на полку {shelf_num}')
            show_all_docs (iterable_1,iterable_2)
    return()  

# Исходные данные имеют следующую структуру:
# перечень всех документов
documents = [
     {'type': 'passport', 'number': '2207 876234', 'name': 'Василий Гупкин'},
     {'type': 'invoice', 'number': '11-2', 'name': 'Геннадий Покемонов'},
     {'type': 'insurance', 'number': '10006', 'name': 'Аристарх Павлов'},
     {'type': 'insurance', 'number': '105', 'name': 'Евсей Ниндзев'}
            ]

# перечень полок, на которых хранятся документы (если документ есть в documents, 
# то он обязательно должен быть и в directories)
directories = {
     '1': ['2207 876234', '11-2','105'],
     '2': ['10006'],
     '3': [],
     '4': []     
               }

inp = ""
while (inp != 'q'):
    inp = input('Введите команду:').lower ()
    if (inp == 'p'):
        inp = input('Введите номер документа: ')
        onwer=show_onwer(documents,inp)
        print(onwer)
# ------------Поиск полки------------------------------
    if (inp == 's'):
        inp = input('Введите номер документа:')
        shelf_num = show_shelf(directories,inp)
        if shelf_num == None:
            print(f'Документ {inp} не найден в базе')
        else:
            print(f'Документ {inp} хранится на полке: {shelf_num}')
# ------------Показать все документы----------------------------------            
    if (inp == 'l'):
        show_all_docs(documents, directories)
#------------Добавить полку------------------------------------------
    if (inp == 'ads'):
        inp = input('Введите номер полки:')
        add_shelf(directories,inp)
#------------Удалить полку------------------------------------------        
    if (inp == 'ds'):
        inp = input('Введите номер полки:')
        del_shelf(directories,inp)
#-------------Добавить документ-------------------------------------
    if (inp == 'ad'):
        add_doc(documents, directories)
#-------------Удалить документ-------------------------------------
    if (inp == 'd'):
        del_doc(documents, directories)
#-------------Переместить документ-------------------------------------
    if (inp == 'm'):
        move_doc(documents, directories)        
        
print ('Работа завершена')

````
