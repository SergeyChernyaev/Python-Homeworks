**Задание**

Смоделировать игру против лотерейного автомата типа “777”. Игрок платит 1 руб., после чего выпадает случайное целое число, равномерно распределенное от 0 до 999. При некоторых значениях числа игрок получает выигрыш.



777: 200 руб.


999: 100 руб.


555: 50 руб.

333: 15 руб.

111: 10 руб.

*77: 5 руб.

**7: 3 руб.

*00: 2 руб.

**0: 1 руб.

**Вопросы:**
1. Выгодна ли игра игроку?
2. Сколько в среднем приобретает или теряет игрок за одну игру?


**Механический игровой автомат с последовательным вращением барабанов:**

```python
%%time
import numpy as np

trials = 10**8 # Количество игр - 100 миллионов
successes = 0    # Количество успехов
balance = 0 # Баланс игрового автомата
win_all = 0 # Общая сумма призовых выплат
for i in range(trials):
    
    balance -=1 # Игрок платит 1 руб.
    win = 0 # выигрыша пока нет
    # Вращается барабан:
    #--------------------Блок вращения по разрядам-----------------------------
    first_num = np.random.randint(0, 10)    # Случайная 1 разряда  от 0 до 9
    second_num = np.random.randint(0, 10)   # Случайная 2 разряда  от 0 до 9
    third_num = np.random.randint(0, 10)    # Случайная 3 разряда  от 0 до 9
    #-----------------------------------------------------------------
       
    if first_num == 7 and second_num == 7 and  third_num == 7: 
        win = 200 
        win_all +=win
    if first_num == 9 and second_num == 9 and  third_num == 9: 
        win = 100
        win_all +=win
    if first_num == 5 and second_num == 5 and  third_num == 5: 
        win = 50   
        win_all +=win
    if first_num == 3 and second_num == 3 and  third_num == 3: 
        win = 15
        win_all +=win
    if first_num == 1 and second_num == 1 and  third_num == 1: 
        win = 10 
        win_all +=win
    if first_num != 7 and second_num == 7 and  third_num == 7: 
        win = 5
        win_all +=win
    if first_num != 7 and second_num != 7 and  third_num == 7: 
        win = 3
        win_all +=win    
        
    if second_num == 0 and third_num == 0: 
        win = 2
        win_all +=win
       
    if second_num != 0 and third_num == 0: 
        win = 1
        win_all +=win 
       
    balance +=win
    
    
   # print (f'Попытка: {i+1} Комбинация:{first_num}{second_num}{third_num} Выигрыш {win} Баланс {balance}')
     
print(f'Количество вращений барабанов:{i+1} Окончательный баланс:{balance} рублей. Общий начисленный выигрыш:{win_all} рублей')        
print(f'Процент отдачи игрового автомата (RTP):{(win_all/trials)*100}')
print(f'Сколько в среднем приобретает (+) или теряет (-) игрок за одну игру: {balance/trials}  рубля')
```

**Электронный игровой автомат с одновременнвм вращением всего барбана:**

```python
%%time

import numpy as np

trials = 10**8 # Количество игр - 100 миллионов
balance = 0 # Баланс игрового автомата
win_all = 0 # Общая сумма призовых выплат
for i in range(trials):
    
    balance -=1 # Игрок платит 1 руб.
    win = 0 # выигрыша пока нет
    # Вращается барабан:
    #--------------------Блок целиком--------------------------------
    num = np.random.randint(0, 1000)    # Случайная полная комбинация
    #-----------------------------------------------------------------
    # ---- Таблица подсчёта выигрышей---------------------------------   
    if num == 777:        win = 200 
    elif num == 999:      win = 100
    elif num == 555:      win = 50   
    elif num == 333:      win = 15
    elif num == 111:      win = 10 
    elif num % 100 == 77: win = 5
    elif num % 10 == 7:   win = 3
    elif num % 100 == 0:  win = 2
    elif num % 10 == 0 :  win = 1
    else: win = 0     
        
    win_all +=win   
    balance +=win
    #print (f'Попытка: {i+1} Число: {num} Комбинация: {num:03} Выигрыш: {win} Баланс: {balance}')
     
print(f'Количество вращений барабанов:{i+1} Окончательный баланс:{balance} рублей. Общий начисленный выигрыш:{win_all} рублей')        
print(f'Процент отдачи игрового автомата (RTP):{(win_all/trials)*100}')
print(f'Сколько в среднем приобретает (+) или теряет (-) игрок за одну игру: {balance/trials}  рубля')
```

**Вывод:** При установленной в задаче цене выигрышных комбинаций выигрыш на стороне игрового автомата.

Игра невыгодна игроку.

**Дополнительное задание повышенной сложности**

Теоретически рассчитать средний выигрыш (проигрыш) и сравнить с результатами моделирования.

```python
# Произведение вероятностей на ценность выпавшей комбинации:

XPay_777 = (1. / 1000) * 200    #  Выигрыш на комбинации 777
XPay_999 = (1. / 1000) * 100    #  Выигрыш на комбинации 999
XPay_555 = (1. / 1000) * 50     #  Выигрыш на комбинации 555
XPay_333 = (1. / 1000) * 15     #  Выигрыш на комбинации 333
XPay_111 = (1. / 1000) * 10     #  Выигрыш на комбинации 111
XPay_77  = (9. / 1000) * 5      #  Выигрыш на комбинации *77
XPay_00  = (10. / 1000) * 2     #  Выигрыш на комбинации *00
XPay_7   = ((45 * 2) / 1000) * 3 # 45 -количество сочетаний из 10 возможных цифр на 2 разряда, поскольку 
                                # наc не интересует уникальность берем и зеркальные комбинации например 12 и 21, 
                                # т.е. умножаем количество сочетаний на 2
XPay_0   = ((45 * 2) / 1000) * 1 # То же самое и для комбиниции **0

X_Pay_all = XPay_777 + XPay_999 + XPay_555 + XPay_333 + XPay_111 + XPay_77 + XPay_00 + XPay_7 + XPay_0
print (f'Средний расчётный выигрыш при ставке в 1 рубль: {X_Pay_all} рубля')
```

**Вывод:** Согласно теории больших чисел при росте числа раундов игры смоделированния отдача автомата приближается к расчётной.
